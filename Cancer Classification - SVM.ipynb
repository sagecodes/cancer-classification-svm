### Training Model

```python
cancer_df.head()
```

```python
X = cancer_df.drop(['target'], axis = 1)
```

```python
X
```

```python
y = cancer_df['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

### Evaluate Model

```python
y_predict = svc_model.predict(X_test)
```

```python
y_predict
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

### improve Model

```python
min_train = X_train.min()
```

```python

cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

Parameter OPtimiation

C paramter contraols trade off between classifying training points correctly and having smooth decision boundary. 

Small c(loose) Makes cost of misclassification low
Large C makes cost of misclassification high. Forcing model to explain data stricter and potentially overfit

Use for trade off of classifiying 

Trial and error to find good c param. 

Gamma conrols how far the influence of a single traing set reaches

large gamma close reach (potential overfit)

small gamma far reach (More generalized)



```python
param_grid = {'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid_predict = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,grid_predict)
```

```python
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, grid_predict))
```
